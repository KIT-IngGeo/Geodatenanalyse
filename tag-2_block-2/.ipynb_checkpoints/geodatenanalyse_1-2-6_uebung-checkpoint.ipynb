## Tag 2 / Block 2 / Übung 6: Zeitreihenanalyse

 In der Datei `GW_Signal.csv` findet Ihr Zeitreihen zur Grundwassertemperatur einiger Pegel aus den vorherigen Übungen. Lest diese ein und schaut Euch die erzeugte Variable an. In der ersten Spalte steht das Datum der Messung, wandelt dieses über `pandas.to_datetime` in den Datentyp `datetime64` um. Geht als Attribut `.format` dafür das spezielle Format der Datumsangabe an ("%d.%m.%Y").

```python
import pandas as pd
signal = pd.read_csv('GW_Signals.csv', sep=';', encoding='cp1252')
signal['Datum']=pd.to_datetime(signal['Datum'], format='%d.%m.%Y')
```

### 1. Visualisierung von Zeitreihen

Stellt nun zuerst alle Zeitreihen gemeinsam in einem Plot dar, und verschafft Euch mit der Abbildung einen Überblick über Trend und Amplituden in den verschiedenen Zeitreihen.  

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(1, figsize=(8, 5))
ax = fig.add_subplot(1,1,1)

for i in np.arange(1, len(signal.columns)):
    plt.plot(signal['Datum'], signal.iloc[:, i])

plt.ylabel("Temperatur [°C]")
plt.xlabel("Datum")
plt.legend(labels = list(signal))
plt.show()
```

### 2. Zerlegung von Zeitreihen

Zerlegt nun eine der Zeitreihen in ihre Komponenten (Trend, Saisonalität und Residuen) mit Hilfe von `statsmodels.api` und der Funktion `tsa.seasonal_decompose()`. Gebt als Attribute die gewünschte Spalte des Datensatzen, das `model` ("additive") und die `period` (365) an. Ausgehend von dem erzeugten Objekt könnt Ihr mit `.plot` die Eregebnisse der Zerlegung graphisch darstellen lassen. 


```python
import statsmodels.api as sm
decomposition = sm.tsa.seasonal_decompose(signal['T108'], model='additive', period =365)
fig = decomposition.plot()
```

Das erzeugte Objekt hat verschiedene Attribute (z.B. trend, seasonal, resid). Erstellt ein Histogramm der Residuen der obigen Zeitreihenanalyse, und interpretiert es im Hinblick auf Normalverteilung, Mittelwert und den Einfluss von Messfehlern. 

```python
residuen = decomposition.resid
fig = plt.hist(residuen)
plt.ylabel("Häufigkeit")
plt.xlabel("Residuen [°C]")
plt.show()
```

Schaut Euch nun als nächstes die Autokorrelation der Zeitreihe an, den sog. "AutoCorrelation Function" (ACF) plot, also die Korrelation eines Wertes in der Zeitreihe mit einem früheren Wert. Importiert dazu `statsmodels.grpahics.tsaplots.plot_acf()` und gebt neben dem Datensatz das Zeitintervall `lags` an. 

Probiert verschiedene Werte für `lags` aus, und interpretiert die Korrelationen in Bezug auf den plot mit Trend und Saisonalität oben. 

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(signal['T108'], lags = 400)
plt.ylabel("Korrelation")
plt.xlabel("Zeitintervall [Tage]")
plt.show()
```

Eine andere beliebte Methode zur Analyse des Trends einer Zeitreihe ist die Bestimmung des gleitenden Mittelwertes (auch moving oder rolling average genannt). Dadurch werden hohe Frequenzen und saisonale Effekte "wegegmittelt" und Zeitreihen geglättet. Für eine `list` geht dies ganz einfach durch Anhängen von `.rolling()` und Definition des Bereichs über den gemittelt werden soll mit dem Argument `window`. Bei `window=10` wird so der arithmetische Mittelwert aus den letzten 10 Messwerten gebildet. 

Testet verschiedene Fensterbreiten und/oder Zeitreihen und vegleicht die ursprüngliche Zeitreihe mit dem Verlauf des gleitenden Mittelwertes. Worin seht Ihr die Vor- und Nachteile von Zerlegung bzw. gleitenden Mittel? 

```python
rolling = signal['T108'].rolling(window=200)
rolling_mean = rolling.mean()
signal['T108'].plot(color='blue')
rolling_mean.plot(color='red')
plt.show()
```

### 4. Regime Shift Analysis

Lest nun den Datensatz "GWT_longterm.csv" ein und schaut ihn im Data Viewer an. Lasst Euch die Zeitreihe als Punkte anzeigen und definiert die Größe der Abbildung dabei so, dass der Verlauf über die Jahre gut sichtbar wird.  


```python
shift= pd.read_csv('GWT_longterm.csv', sep=';', encoding='cp1252')

fig = plt.figure(1, figsize=(15, 5))
ax = fig.add_subplot(1,1,1)

plt.scatter(shift['year'], shift['GWT'], c='b')

plt.ylabel("Grundwassertemperatur")
plt.xlabel("Jahr")
plt.show()
```

In dem Datensatz gibt es einige Datenlücken (NaN), die die weitere Auswertung stören könnten. Eine einfache Option zum Füllen der Lücken ist die Funktion `pandas.DataFrame.interpolate()`. Wendet diese mit dem Argument "linear" auf den Datensatz an, und schaut Euch das Ergebniss an. Wie werden die Lücken hier gefüllt?  

```python
shift = shift.interpolate('linear')
```

Ein genereller Trend mit ansteigenden Temperaturen ist in der Zeitreihen vorhanden. Allerdings ist dieser nicht gleichmäßig (vgl. Grundwassser 1970-1990 und 1990-2020), daher bietet sich hier eine Analyse auf verschiedene Regime in der Zeitreihe an. 

Installiert dafür zuerst das Python Package 'ruptures'. Öffnet Eure Anaconda Konsole und gebt den Befehl "pip install ruptures" ein. Nach einem Augenblick sollte die Konsole die erfolgreiche Installation anzeigen. 



```python
import ruptures as rpt

n = len(shift)  # get length of time series
sigma = 0.1  # estimate for variance in data
GWT = pd.Series.to_numpy(shift['GWT']) # convert pd.Series to np.array

model2 = rpt.BottomUp(model="rbf", min_size=10).fit(GWT) # min_size = minimum length of one regime
shifts_GWT = model2.predict(pen=np.log(n)*sigma**2) # pen = penalty term to avoid getting short and insignificant regimes 
```

Die `list` der Modelvorhersagen enthält die Werte der Indizes derjenigen Zellen im Datensatz, in denen ein Regimewechsel identfiziert wurde. Mit diesen Angaben könnt Ihr über eine Schleife die statistische Eigenschaften (Mittelwert, Varianz, usw.) der einzelnen Regime ermitteln. 

```python
one = np.ones(1, dtype='int')  # create array containing a single 1
regime_GWT = np.append(one, shifts_GWT).tolist() # append array with shift indices to 1, and convert to list

mean_GWT = np.zeros((len(shifts_GWT), 1)) # create array with zeros before loop
var_GWT = np.zeros((len(shifts_GWT), 1))
for i in range(len(shifts_GWT)): # looping over the list with regime indices
    mean_GWT[i] = np.median(GWT[regime_GWT[i]:regime_GWT[i+1]:1]) # mean of cells in each regime
    var_GWT[i] = np.var(GWT[regime_GWT[i]:regime_GWT[i+1]:1]) # variance of cells in each regime 
```

Als letztes könnt Ihr nun noch die Zeitreihen zusammen mit den identifzierten Regimen und deren Mittelwerten darstellen 

```python
fig = plt.figure(1, figsize=(15, 5))
ax = fig.add_subplot(1,1,1)

plt.scatter(shift['year'], shift['GWT'], c='b')

for i in range(len(shifts_GWT)):
    line = plt.plot([regime_GWT[i]+min(shift['year']), regime_GWT[i+1]+min(shift['year'])],
                    [mean_GWT[i], mean_GWT[i]], '--', c='k', lw=1)

ax.set_xlim(min(shift['year']), max(shift['year']))
ax.set_ylabel('Grundwassertemperatur[°C]')
ax.set_xlabel('Jahre')
plt.show()
```
