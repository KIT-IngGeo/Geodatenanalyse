### Objekte der Abbildung

**Figure**: Es ist eine ganze Abbildung, die eine oder mehrere Achsen (Plots) enthalten kann. Sie können sich eine Abbildung als eine Leinwand vorstellen, die Plots enthält.


**Axes**: Das ist das, was man sich im Allgemeinen unter einem Plot vorstellt. Eine Abbildung kann viele Achsen enthalten. Sie enthält zwei oder drei (im Falle von 3D) Axis-Objekte. Jede Achse hat einen Titel, eine x-Beschriftung und eine y-Beschriftung.

**Axis**: Sie sind die nummernstrichähnlichen Objekte und kümmern sich um die Erzeugung der Diagrammgrenzen.

**Artist**: Alles, was man auf der Figur sehen kann, ist ein *Artist*, wie z.B. Text-Objekte, Line2D-Objekte, Collection-Objekte. Die meisten *Artist* sind an Achsen gebunden.

### Erste Schritte mit Pyplot

Pyplot ist ein Modul von Matplotlib, das einfache Funktionen bereitstellt, um Plot-Elemente wie Linien, Bilder, Text usw. zu den aktuellen Achsen in der aktuellen Abbildung hinzuzufügen.

Bevor wir eine einfache Abbildung machen, müssen wir das Paket laden:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# alle x-Werte
x = [0,1,2,3,4,5]

# alle y-Werte
y = [1,1,2.5,3,5,4]

plt.plot(x, y)
```

So einfach ist das!

**Achtung**: Die Variablen sind hier noch in einer Liste! Am besten werden *numpy* Vektoren verwendet.

## Abbildungsgröße einstellen

Meistens wollen wir die Abbildungsgröße bestimmen. Hierzu müssen wir die Methode *figure()* mit keyword *figsize* verwenden.

```python
plt.figure(figsize=[8,4])
plt.plot(x, y)
```

**Achtung**: *figsize* ist hier in Inches! Bei Bedarf muss das umgerechnet werden.

### Einfache Einstellung der Dateneigenschaften

Die Methode *plot* hat eine flexible Erweiterung, welche über Kürzel angesprochen werden kann.

```python
plt.plot(x, y, '.r')
```

Hier steht jedes Zeichen für ein Kürzel. Eine Liste der Kürzel findet ihr hier: [Kürzel](https://matplotlib.org/api/pyplot_api.html)

Man kann auch mehrer Datensätze hinzufügen:

```python
plt.plot(x, y, '-b')
plt.plot(x, y, '.r')
```

### Die Legende

Wenn mehrere Datensätze abgebildet werden, sollte man eine Legende zur Orientierung verwenden.

Die einfachste Möglichkeit hierfür ist das Schlüsselwort *label*:

```python
# siehe Label hier
plt.plot(x, y, '-b', label='Daten 1')
plt.plot(x, y, '.r', label='Daten 2')

# Legende hinzufügen
plt.legend()
```

### Der Abbildungstitel

```python
# siehe Label hier
plt.plot(x, y, '-b', label='Daten 1')
plt.plot(x, y, '.r', label='Daten 2')

# Legende hinzufügen
plt.legend()

# der Titel
plt.title('Eine schöne Grafik')
```

## Abbildung als Datei speichern

Erstellte Abbildungen können ganz einfach als Datei abgespeichert werden.

**Pixelgrafik**: Eine Rastergrafik (auch Pixelgrafik) ist eine Form der Beschreibung eines Bildes in Form von computerlesbaren Daten. Rastergrafiken bestehen aus einer rasterförmigen Anordnung sogenannter Pixel, denen jeweils eine Farbe zugeordnet ist.
Formate: JPG, PNG, GIF, BMP, etc.

Quelle: [Wikipedia](https://de.wikipedia.org/wiki/Rastergrafik)

**Vektorgrafik**: Eine Vektorgrafik ist eine Computergrafik, die aus grafischen Primitiven wie Linien, Kreisen, Polygonen oder allgemeinen Kurven (Splines) zusammengesetzt ist. Meist sind mit Vektorgrafiken Darstellungen gemeint, deren Primitive sich zweidimensional in der Ebene beschreiben lassen. Formate: PDF, EPS, PS, etc.

Quelle: [Wikipedia](https://de.wikipedia.org/wiki/Vektorgrafik)

```python
plt.plot(x, y, '-b')
plt.plot(x, y, '.r')

plt.xlabel('Die Zahl Pi')
plt.ylabel('Die Sinus-Funktion')

# speichern als Pixelgrafik
plt.savefig('Abbildung.png', dpi=150)
```

```python
plt.plot(x, y, '-b')
plt.plot(x, y, '.r')

plt.xlabel('Die Zahl Pi')
plt.ylabel('Die Sinus-Funktion')

# speichern als Vektorgrafik
plt.savefig('Abbildung.pdf')
```

## Konfiguration von Abbildungen und Achsenobjekten

Für ein gutes Verständnis von Abbildungen, sollte man die Elemente gut verstehen.

Dazu gehört eine gute Kontrolle der einzelnen Attribute.

### Einfache Einstellung der Datenzeiger

Dies ist die Position, an welchen die kleinen Striche zur Orientierung angezeigt werden. 

Normalerweise wird dieses automatisch erstellt.

Aber machmal muss oder möchte man eingreifen:

```python
# Variablen setzen
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

# Abbildung erstellen
plt.plot(x, y)
```

```python
# Abbildung erstellen
plt.plot(x, y)
xticks = plt.xticks()
print(xticks)
```

```python
# Abbildung erstellen
plt.plot(x, y)

# neu setzen
plt.xticks([1,2,5,7,10.5])
```

```python
# Abbildung erstellen
plt.plot(x, y)

# Rster aktivieren
plt.grid()
```

### Einfache Achsenbeschriftung

Die Achsen sind standardmäßig nicht beschriftet.

Das lässt sich aber leicht ändern.

```python
# Abbildung erstellen
plt.plot(x, y)

# x-Achse beschriften
plt.xlabel('Das ist unten')

# y-Achse beschriften
plt.ylabel('Das ist vertikal')
```

### Die Abbildung als Objekt

```python
# eine Abbildung erstellen
fig = plt.figure()

# eine Achse hinzufügen
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Abbildung erstellen
ax.plot(x, y)
```

**Bemerkung**: Jetzt haben wir ein Achsenobjekt *ax* und können diese ändern!

### Achsenbeschriftung

Hier gibt es folgende Hauptelemente:
- Datenzeiger (ticks) und Datenbeschriftung (ticklabels)
- Datenbgrenzung (limits)
- Achsenbeschriftung (labels)

### Achsenzeiger (ticks) und Achsenzeigerbeschriftung (ticklabels)

Dies ist die Beschriftung der kleinen Striche an den Achsen. 

Normalerweise wird dieses automatisch erstellt.

Aber machmal muss man eingreifen:

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)

# Datenzeiger und -beschriftung ändern
ax.set_xticks([0,np.pi/2,np.pi,2*np.pi])
ax.set_xticklabels(['0','pi/2','pi','2*pi'])
```

### Datenbgrenzung (limits)

Die Begrenzung sind die x/y-Werte, an welchen die Achsen anfangen und aufhören.

Normalerweise werden diese automatisch eingestellt.

Aber machmal muss man eingreifen:

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)

# Begrenzung der x-Achse ändern
ax.set_xlim([2,3.5])
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)

# Begrenzung der y-Achse ändern
ax.set_ylim([-0.5,0.5])
```

## Abbildungstypen: Linien und Punkte (Daten)

Linien und Punkte hatten wir bereits gezeigt (siehe weiter oben).

Wann sollte man welche Abbildungsart nehmen?

- **Punkte**: Für *diskrete* Daten, oder auch geringe Datendichte.
- **Linien**: Für *kontinuierliche* Daten, oder auch hohe diskrete Datendichte.

Diese Abbildungstypen sind für begrenzte Datenmengen geeignet.

```python
# geringe Datendichte
x1 = np.linspace(0, 4*np.pi, 10)
y1 = np.sin(x1)

# hohe Datendichte
x2 = np.linspace(0, 4*np.pi, 100)
y2 = np.sin(x2)

# Abbildung erstellen
plt.plot(x2, y2, '-', c=[0.7,0.7,0.7])
plt.plot(x1, y1, '-k')
plt.plot(x1, y1, '.r')
```

## Abbildungstyp Balkendiagramm

Für Vergleiche oder große Datenmengen sollte man andere Abbildungsarten nehmen.

Das **Balken-/Säulendiagramm** eignet sich besonders, um wenige Ausprägungen (bis ca. 15) zu veranschaulichen.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# die Daten
semester = ['1', '2', '3', '4']
students = [23, 17, 35, 29]

# Balkendiagramm
ax.bar(semester, students)

ax.set_xlabel('Semester')
ax.set_ylabel('Anzahl Studiernde')
```

```python
# die Daten
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)

# the x locations for the groups
index = np.arange(1, 6) 
width = 0.35

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# gestaffelte Balken
ax.bar(index, menMeans, width, color='r')
ax.bar(index, womenMeans, width, bottom=menMeans, color='b')

# Beschriftungen
ax.set_ylabel('Scores')

# Achsenzeigerbeschriftungen
ax.set_xticks(index, ('G1', 'G2', 'G3', 'G4', 'G5'))
ax.set_yticks(np.arange(0, 81, 10))

# Legende
ax.legend(labels=['Men', 'Women'])
```

## Abbildungstyp Box-Plot (für Statistiken)

Die **Box-Plot/Kastengrafik** ist ein Diagramm, das zur grafischen Darstellung der Verteilung eines mindestens ordinalskalierten Merkmals verwendet wird.

Boxplot einer Normalverteilung:

Quelle: [Towards Datascience](https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51)

    

```python
# Daten erzeugen
data = np.random.rand(1000)

# Box-Plot erstellen
plt.boxplot(data)
```

Was ist mit "Ausbrechern" (outliers)? 

Lass uns dies mal simulieren:

```python
# erstelle zufällige Daten
sampl = np.random.uniform(0.5, 13.3, 500)
flier_high = np.random.rand(10) * 20 + 20
flier_low = np.random.rand(10) * -10

# fasse die Daten in einem Vektor zusammen
data = np.concatenate((sampl, flier_high, flier_low))

# Darstellung ...
fig, ax = plt.subplots()
ax.boxplot(data)
plt.show()
```

Wir können das Box-Plot auch noch genauer anpassen:

```python
# erstelle zufällige Daten
sampl = np.random.uniform(0.5, 13.3, 500)
flier_high = np.random.rand(10) * 20 + 20
flier_low = np.random.rand(10) * -10

# fasse die Daten in einem Vektor zusammen
data = np.concatenate((sampl, flier_high, flier_low))

# Darstellung ...
fig, ax = plt.subplots(1)

# Anpassung der Datenpunkte, Orientierung und Breite
red_square = dict(markerfacecolor='r', marker='o')
ax.boxplot(data, vert=False, flierprops=red_square, whis=0.9)
plt.show()
```
